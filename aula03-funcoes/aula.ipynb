### Exemplo


$Resumo$
É um gráfico que mostra os batimentos do coração com zoom entre 15 e 16 segundos e com os valores de -1 a 3. <br>

$Explicações$ <br>

- electrocardiogram() - devolve sinais de batimentos cardiacos <br>
- fs - frequencia de amostragem (360 valores por segundo) <br>
- time - tempo do grafico <br>
- ecg.size - qtd total de valores do egc <br>
- np.arange(ecg.size) - cria sequencia de 0 a egc <br>
    - ao dividir por fs transforma posições em segundos <br>
    - 0 / 360 = 0s, 360/360 = 1s, 720/360 = 2s <br>
- plt.plot(x, y) - gráfico é desenhado <br>
- plt.plot(x, y) - x horizontal, y vertical <br>
- plt.xlabel() - legenda <br>
- plt.xlim(15, 16) - zoom no grafico, mostra so entre 15 e 16 segs <br>
- plt.ylim(-1, 3) - limita altura do grafico <br>
- plt.show() - mostra o grafico <br>

```python
from scipy.datasets import electrocardiogram
ecg = electrocardiogram()

#
fs = 360
time = np.arange(ecg.size) / fs
plt.plot(time, ecg)
plt.xlabel("Tempo [s]")
plt.ylabel("Sinal de ECG [mV]")
plt.xlim(15, 16)
plt.ylim(-1, 3)
plt.show()
```

```python
ecg.ndim
```

### Exemplo

Image.open() - abre a imagem. <br>
plt.imshow() - prepara/desenha ela <br>
plt.show() - mostra na tela <br>

```python
# PIL abre, edita, converte e salva imagens
from PIL import Image

# carregar a imagem local e converte para tons de cinza (L)
img = Image.open("imgFIAP.png").convert("L")
img_array = np.array(img)   # transforma em array para modificar
```

```python
# largura e altura da img/gráfico
plt.figure(figsize=(8, 6))

# imshow - abre e desenha a img no grafico
# cmap="gray" - mapa de cores "escala de cinza"
plt.imshow(img_array, cmap="gray")

# titulo da img
plt.title("Imagem Preto e Branco")
plt.show()
```

```python
type(img_array)
```

```python
img_array.ndim
```

```python
img_array.shape
```

### Zoom no exemplo

```python
x0, y0 = 100, 50   # canto superior esquerdo da região (linha, coluna)
w, h = 10, 10       # largura e altura do quadrado
```

Aqui ta recortando a img, é como se rolasse: <br>
linha (y0:y0+h) de 104 a 114 (+10 do h), coluna ((x0:x0+w)) de 440 a 450 (+10 do w) <br>
O resultado é uma mini img 10x10

```python
# recorta um pedaço da img
crop = img_array[y0:y0+h, x0:x0+w]   # sub-região
```

```python
plt.figure(figsize=(6, 6))

plt.imshow(crop, cmap="gray")
plt.title(f"Sub-região {w}×{h}")

# mostra os numeros dos eixos x e y
plt.xticks(range(w))
plt.yticks(range(h))

plt.show()
```

### Função personalizada

Retorna os elementos escolhidos de x ou y, dependendo da condição.
<br>
$sintase: np.where(condição, valor_se_true, valor_se_false)$

```python
def my_abs3(x):
    return np.where(x >= 3, -x, x)
```

```python
print(my_abs3(a))
```
